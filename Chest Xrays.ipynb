### Check For Data Leakage

```python
def check_for_data_leakage(df1, df2, col):
    '''Returns False if no overlapping of Patient IDs between 2 dataframes'''
    unique_df1 = set(df1[col].values)
    unique_df2 = set(df2[col].values)
    
    overlapping_patients = unique_df1.intersection(unique_df2)
    
    leakage = len(overlapping_patients)>0
    
    return leakage
```

```python
print("leakage between train and test: {}".format(check_for_data_leakage(train, test, 'PatientId')))
print("leakage between valid and test: {}".format(check_for_data_leakage(valid, test, 'PatientId')))
```

### Preparing Images

```python
def get_train_generator(df, x_col, y_cols, target_sz, img_dir, batch_size=8, shuffle=True, seed=1):
    "returns a train generator with normalized batches"
    
    image_gen = ImageDataGenerator(samplewise_center=True,
                                   samplewise_std_normalization=True,
                                  rotation_range = 5,
                                  width_shift_range=0.1,
                                   height_shift_range=0.1,
                                   shear_range=0.1,
                                   zoom_range=0.1,
                                   fill_mode="nearest",
                                   cval=0.0,
                                   rescale = 0.1)
    
    train_gen = image_gen.flow_from_dataframe(dataframe=df,
                                              directory=img_dir,
                                              x_col=x_col,
                                              y_col=y_cols,
                                             class_mode='raw',
                                             batch_size=batch_size,
                                             shuffle=shuffle,
                                             seed=seed,
                                             target_size=target_sz)
    
    return train_gen
```

We cannot use the same generator for validation and test sets 

```python
def get_val_and_test_generator(train_df, valid_df, test_df, x_col, y_cols,img_dir, target_sz,
                               batch_size=8,sample=100,seed=1):
    '''Return valid and test generators with normazlized images with statisitcs computed from training set'''
    
    raw_train_generator = ImageDataGenerator().flow_from_dataframe(dataframe=train_df,
                                                                  directory=img_dir,
                                                                   x_col=x_col, 
                                                                   y_col=y_cols, 
                                                                   class_mode="raw", 
                                                                   batch_size=sample, 
                                                                   shuffle=True, 
                                                                   target_size=target_sz)
    batch = raw_train_generator.next()
    data_sample = batch[0] #shape of sample, target_sz, nc (100,1024,1024,3)
    
    image_generator = ImageDataGenerator(featurewise_center=True,
                                        featurewise_std_normalization=True)
    
    image_generator.fit(data_sample)
    
    val_generator = image_generator.flow_from_dataframe(dataframe=valid_df,
            directory=image_dir,
            x_col=x_col,
            y_col=y_cols,
            class_mode="raw",
            batch_size=batch_size,
            shuffle=False,
            seed=seed,
            target_size=target_sz)

    test_generator = image_generator.flow_from_dataframe(
            dataframe=test_df,
            directory=image_dir,
            x_col=x_col,
            y_col=y_cols,
            class_mode="raw",
            batch_size=batch_size,
            shuffle=False,
            seed=seed,
            target_size=target_sz)
    
    return val_generator, test_generator
```

```python
image_dir = 'images-small/'
train_generator = get_train_generator(train, 'Image', labels,(320,320),image_dir)
valid_generator, test_generator= get_val_and_test_generator(train, valid, test, "Image", labels,image_dir,(320,320))
```

# Model Development

### Addressing class imbalance issue

```python
plt.xticks(rotation=90)
plt.bar(x=labels, height=np.mean(train_generator.labels, axis=0))
plt.title("Frequency of Each Class")
plt.show()
```

Using weighted loss to handle class imbalance

```python
def compute_class_freq(labels):
    n = labels.shape[0]
    
    pos_freq = np.sum(labels, axis=0)/n
    neg_freq = 1 - pos_freq
    
    return pos_freq, neg_freq
```

```python
pos_weights, neg_weights = compute_class_freq(train_generator.labels)
```

```python
def get_weighted_loss(pos_weights, neg_weights, epsilon=1e-7):
    """
    Return weighted loss function given negative weights and positive weights."""
    def weighted_loss(y_true, y_pred):
        loss = 0.0

        for i in range(len(pos_weights)):
            loss += -1 * K.mean((pos_weights[i] * y_true[:,i] * K.log(y_pred[:,i] + epsilon)
                               + neg_weights[i] * (1 - y_true[:,i]) * K.log(1 - y_pred[:,i] + epsilon)))
        return loss
    
    return weighted_loss
```

first we train on a new model to set as a baseline

```python
def Eff_B7_NS():
    model_EfficientNetB7_NS = Sequential([efn.EfficientNetB7(input_shape=(320,320,3),include_top=False),
                                 GlobalAveragePooling2D(),
                                          Dense(len(labels), activation="sigmoid")])
    model_EfficientNetB7_NS.compile(optimizer='Adam',loss = 'categorical_crossentropy',metrics=['categorical_accuracy'])
    
    
    return model_EfficientNetB7_NS

def Inc_Res_V2():
    model_EfficientNetB7_NS = Sequential([InceptionResNetV2(input_shape=(320,320,3),include_top=False),
                                 GlobalAveragePooling2D(),
                                          Dense(len(labels), activation="sigmoid")])
    model_EfficientNetB7_NS.compile(optimizer='Adam',loss = 'categorical_crossentropy',metrics=['categorical_accuracy'])
    
    
    return model_EfficientNetB7_NS
```

```python
base_model = DenseNet121(include_top=False)

x = base_model.output

x = GlobalAveragePooling2D()(x)
predictions = Dense(len(labels), activation="sigmoid")(x)

model = Model(inputs=base_model.input, outputs=predictions)
model.compile(optimizer='adam', loss=get_weighted_loss(pos_weights, neg_weights))
```

## Predicition

```python
predicted_vals = model.predict(test_generator, steps = len(test_generator))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
def get_roc_curve(labels, predicted_vals, generator):
    auc_roc_vals = []
    for i in range(len(labels)):
        try:
            gt = generator.labels[:, i]
            pred = predicted_vals[:, i]
            auc_roc = roc_auc_score(gt, pred)
            auc_roc_vals.append(auc_roc)
            fpr_rf, tpr_rf, _ = roc_curve(gt, pred)
            plt.figure(1, figsize=(10, 10))
            plt.plot([0, 1], [0, 1], 'k--')
            plt.plot(fpr_rf, tpr_rf,
                     label=labels[i] + " (" + str(round(auc_roc, 3)) + ")")
            plt.xlabel('False positive rate')
            plt.ylabel('True positive rate')
            plt.title('ROC curve')
            plt.legend(loc='best')
        except:
            print(
                f"Error in generating ROC curve for {labels[i]}. "
                f"Dataset lacks enough examples."
            )
    plt.show()
    return auc_roc_vals
```
